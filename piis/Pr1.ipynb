### Задача №1
Написать функцию, на вход которой подается строка, состоящая из латинских букв. Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке.

```python
def count_vowels(str):
  k = 0
  vowels = "aeiouAEIOU"
  for i in str:
    if i in vowels:
      k += 1

  return k

count_vowels(input())
```

```python
def count_bit(n):
  if n <= 0:
    raise ValueError("The number must be a positive")

  return bin(n).count('1')

count_bit(int(input()))
```

### Задача №4
Написать функцию, которая принимает положительное число. Функция должна вернуть то, сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы получилось число, состоящее из одной цифры.
Например, для входного числа:
- 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
- 4 функция должна вернуть 0, так как число уже состоит из одной цифры
- 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 => 1*2=2


```python
def count_multi(n):
  if n < 10:
    return 0

  k = 0
  while n >= 10:
    multi = 1
    while n > 0:
      multi *= n % 10
      n //= 10
    n = multi
    k += 1

  return k

count_multi(int(input()))
```

### Задача №5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и возвращает среднеквадратическое отклонение двух векторов.

```python
def std(vec1, vec2):
  if len(vec1) != len(vec2):
    raise ValueError("The lenght must be the same")
  n = len(vec1)
  M1 = sum(vec1) / n
  M2 = sum(vec2) / n
  D1 = 0
  D2 = 0
  for i in range(n):
    D1 += ((vec1[i] - M1) ** 2) / n
    D2 += ((vec2[i] - M2) ** 2) / n
  std1 = D1 ** 0.5
  std2 = D2 ** 0.5

  return std1, std2

std([1, 2, 3, 4, 5], [6, 10, 7, 5, 9])
```

### Задача №6
Написать функцию, которая принимает список чисел и возвращает кортеж из двух элементов. Первый элемент кортежа – мат. ожидание, второй элемент – СКО. Запрещается использовать функции для расчета соответствующих характеристик.

```python
def mean_std(a):
  n = len(a)
  M = sum(a) / n

  D = 0
  for i in a:
    D += (i - M) ** 2 / n
  std = D ** 0.5

  return (round(M, 2), round(std, 2))

l = list([1, 3, 3, 6, 8])
mean_std(l)
```

### Задача №7
Написать функцию, принимающая целое положительное число. Функция должна вернуть строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”

```python
def prime(n):
  if n <= 0:
    raise ValueError("The number must be a positive integer")

  multiplier = {}
  d = 2
  while n > 1:
    while n % d == 0:
      if d in multiplier:
        multiplier[d] += 1
      else:
        multiplier[d] = 1
      n //= d
    d += 1

  res = []
  for prime, power in multiplier.items():
    if power == 1:
      res.append(f"({prime})")
    else:
      res.append(f"({prime}**{power})")

  return ''.join(res)

prime(int(input()))
```

### Задача №8
Написать функцию, принимающая 2 строки вида “xxx.xxx.xxx.xxx” представляющие ip-адрес и маску сети. Функция должна вернуть 2 строки: адрес сети и широковещательный адрес.

```python
def network_broadcast(ip, mask):
  ip_bin = ''.join([bin(int(x) + 256)[3:] for x in ip.split('.')])
  mask_bin = ''.join([bin(int(x) + 256)[3:] for x in mask.split('.')])

  network_bin = ''.join([str(int(ip_bin[i]) & int(mask_bin[i])) for i in range(32)])
  network = '.'.join([str(int(network_bin[i:i+8], 2)) for i in range(0, 32, 8)])

  broadcast_bin = ''.join([str(int(ip_bin[i]) | (1 - int(mask_bin[i]))) for i in range(32)])
  broadcast = '.'.join([str(int(broadcast_bin[i:i+8], 2)) for i in range(0, 32, 8)])

  return network, broadcast

network_broadcast("192.168.1.1", "255.255.255.0")
```

### Задача №9
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна вернуть k, иначе строку “It is impossible”.

```python
def is_possible(n):
  k = 0
  sum_square = 0
  while sum_square < n:
    k += 1
    sum_square += k * k
  if sum_square == n:
    return k
  else:
    return "It is impossible"

is_possible(int(input()))
```

### Задача №10
Написать функцию, которая принимает положительное целое число n и определяющая является ли число n сбалансированным. Число является сбалансированным, если сумма цифр до средних цифр равна сумме цифр после средней цифры. Если число нечетное, то средняя цифра одна, если четное, то средних цифр две. При расчете, средние числа не участвуют.
Например:
- Число 23441 сбалансированное, так как 2+3=4+1
- Число 7 сбалансированное, так как 0=0
- Число 1231 сбалансированное, так как 1=1
- Число 123456 несбалансированное, так как 1+2!=5+6


```python
def is_balanced(n):
  if n <= 0:
    raise ValueError("The number must be a positive integer")

  digits = [int(i) for i in str(n)]
  mid = len(digits) // 2

  if len(digits) % 2 == 0:
    sum_left = sum(digits[:mid])
    sum_right = sum(digits[mid:])
  else:
    sum_left = sum(digits[:mid])
    sum_right = sum(digits[mid+1:])

  return sum_left == sum_right

is_balanced(int(input()))
```

### Задача №11
Написать функцию, которая принимает двумерный массив M в первой колонке которой стоит буква латинского алфавита (данная буква обозначает принадлежность к классу) и вещественное число r, которое удовлетворяет условию 0<r<1. Функция должна разбить данный массив на 2 так, что количество строк в новых массивах пропорционально r и (1-r). Также, в каждом новом массиве, количество строк одного класса должно быть пропорционально количеству строк этого класса в исходном массиве.

```python
def split_array(M, r):
  if not (0 < r < 1):
    raise ValueError("0 < r < 1")

  # Количество строк каждого класса в исходном массиве
  class_counts = {}
  for row in M:
    cls = row[0]
    if cls in class_counts:
      class_counts[cls] += 1
    else:
      class_counts[cls] = 1

  rows_count1 = {cls: int(count * r) for cls, count in class_counts.items()}

  first_array = []
  second_array = []

  for cls in class_counts:
    class_rows = [row for row in M if row[0] == cls]
    first_array.extend(class_rows[:rows_count1[cls]])
    second_array.extend(class_rows[rows_count1[cls]:])

  return first_array, second_array

# Пример использования
M = [
  ['a', 1, 2],
  ['a', 3, 4],
  ['b', 5, 6],
  ['b', 7, 8],
  ['c', 9, 10],
  ['c', 11, 12]
]

r = 0.5

first_array, second_array = split_array(M, r)
print("First array:")
for row in first_array:
  print(row)
print("\nSecond array:")
for row in second_array:
  print(row)
```
